```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
def season(month):
    if month>=3 and month<=5: return 0
    elif month>=6 and month<=8: return 1
    elif month>=9 and month<=11: return 2
    else: return 3
```

```python
weather = pd.read_csv('./data/weather.csv')
weather.head(5)
```

```python
weather['일시'] = pd.to_datetime(weather['일시'].astype(str), format='%Y-%m-%d %H:%M')
weather['일시'] = weather['일시'].dt.strftime('%Y%m%d %H')

weather.head(5)
```

```python
tashu = pd.read_csv('./data/tashu.csv')
tashu = tashu.dropna(axis=0)

tashu = tashu[tashu.RENT_DATE != 0]
tashu = tashu[tashu.RETURN_DATE != 0]
tashu = tashu[tashu.RENT_STATION != 0]
tashu = tashu[tashu.RETURN_STATION != 0]

tashu['RENT_DATE'] = pd.to_datetime(tashu['RENT_DATE'].astype(str), format='%Y%m%d%H%M%S')
tashu['RENT_DATE'] = tashu['RENT_DATE'].dt.strftime('%Y%m%d %H')

tashu['RETURN_DATE'] = pd.to_datetime(tashu['RETURN_DATE'].astype(str), format='%Y%m%d%H%M%S')
tashu['RETURN_DATE'] = tashu['RETURN_DATE'].dt.strftime('%Y%m%d %H')

tashu.head(10)
```

```python
tashu_rent=pd.DataFrame(columns=['RENT_STATION', 'RENT_DATE'])
tashu_return=pd.DataFrame(columns=['RETURN_STATION', 'RETURN_DATE'])

tashu_rent['RENT_STATION'] = tashu["RENT_STATION"]
tashu_rent['RENT_DATE'] = tashu['RENT_DATE']
tashu_rent = tashu_rent.sort_values(["RENT_STATION"], ascending=[True])

tashu_return['RETURN_STATION'] = tashu["RETURN_STATION"]
tashu_return['RETURN_DATE'] = tashu['RETURN_DATE']
tashu_return = tashu_return.sort_values(["RETURN_STATION"], ascending=[True])

tashu_rent.head(5)
```

```python
tashu_return.head(5)
```

```python
station = pd.read_csv('./data/station.csv')
station.head(5)
```

```python
tmp_data = pd.DataFrame(columns=['STATION', 'YEAR', 'MONTH', 'DAY', 'HOUR', 'WEEKDAY', 'SEASON', 'TEMPERATURE', 'RAINFALL', 'WINDSPEED'
                                , 'HUMLDITY', 'SNOWFALL'])

j = 0

for i in range(2):
    start = datetime(2013+i, 1, 1, 00)
    end =datetime(2013+i, 12, 31, 23)

    tmp = start+timedelta(hours=24)
    while start <= end:
        month = str(start.month)
        year = str(start.year)
        day = str(start.day)
        hour = str(start.hour)
    
        if len(month) < 2:
            month = "0"+month
        
        if len(day) < 2:
            day = "0"+day
        
        if len(hour) < 2:
            hour = "0"+hour
        
        dt = datetime(int(year), int(month), int(day))
        wt = weather[(weather['일시'] == year+month+day+" "+hour)]

        wd = dt.weekday()     
        tmp_data.loc[j] = ('1', start.year, start.month, start.day, start.hour, wd, season(int(month)), wt['기온(°C)'].values[0], 
                       wt['강수량(mm)'].values[0], wt['풍속(m/s)'].values[0], wt['습도(%)'].values[0], wt['적설(cm)'].values[0])
    
        j = j+1
        start = start+timedelta(hours=1)
        
tmp_data = tmp_data.fillna(0)
tmp_data.head(10)
```

```python
tmp_data["STATION"] = 1
total_rent=pd.DataFrame(columns=['STATION', 'YEAR', 'MONTH', 'DAY', 'HOUR', 'WEEKDAY', 'SEASON', 'TEMPERATURE', 'RAINFALL', 'WINDSPEED'
                                , 'HUMLDITY', 'SNOWFALL'])

total_return=pd.DataFrame(columns=['STATION', 'YEAR', 'MONTH', 'DAY', 'HOUR', 'WEEKDAY', 'SEASON', 'TEMPERATURE', 'RAINFALL', 'WINDSPEED'
                                , 'HUMLDITY', 'SNOWFALL'])

total_rent = pd.concat([total_rent, tmp_data])
total_rent.head(5)
```

```python
for i in range(1, len(station['번호'])):
    tmp_data["STATION"] = station['번호'][i]
    total_rent = pd.concat([total_rent, tmp_data])
 
```

```python
total_return = total_rent
total_return.head(5)
```

```python
rent_count = tashu_rent.groupby(["RENT_STATION", "RENT_DATE"])["RENT_STATION"].count().reset_index(name="RENTCOUNT")
return_count = tashu_return.groupby(["RETURN_STATION", "RETURN_DATE"])["RETURN_STATION"].count().reset_index(name="RETURNCOUNT")
rent_count.head(10)
```

```python
return_count.head(10)
```

```python
# RentCount 
rent_count['YEAR'] = rent_count['RENT_DATE'].apply(lambda e: int(e[0:4]))
rent_count['MONTH'] = rent_count['RENT_DATE'].apply(lambda e: int(e[4:6]))
rent_count['DAY'] = rent_count['RENT_DATE'].apply(lambda e: int(e[6:8]))
rent_count['HOUR'] = rent_count['RENT_DATE'].apply(lambda e: int(e.split()[-1]))

rent_count = rent_count.drop('RENT_DATE', axis=1)
rent_count = rent_count.rename(columns = {'RENT_STATION': 'STATION'})

rent_count["STATION"] = pd.to_numeric(rent_count["STATION"])
#rent_count.to_csv('./data/2014_rent_count.csv')
```

```python
# ReturnCount
return_count['YEAR'] = return_count['RETURN_DATE'].apply(lambda e: int(e[0:4]))
return_count['MONTH'] = return_count['RETURN_DATE'].apply(lambda e: int(e[4:6]))
return_count['DAY'] = return_count['RETURN_DATE'].apply(lambda e: int(e[6:8]))
return_count['HOUR'] = return_count['RETURN_DATE'].apply(lambda e: int(e.split()[-1]))

return_count = return_count.drop('RETURN_DATE', axis=1)
return_count = return_count.rename(columns = {'RETURN_STATION': 'STATION'})

return_count["STATION"] = pd.to_numeric(return_count["STATION"])
#return_count.to_csv('./data/2014_return_count.csv'
```

```python
merge_rent = pd.merge(total_rent, rent_count, on=['STATION', 'YEAR', 'MONTH', 'DAY', 'HOUR'], how="left" )
merge_rent = merge_rent.fillna(0)

merge_rent.head(10)
merge_rent.to_csv('./data/2013-2014_rent.csv', index=False)
```

```python
merge_return = pd.merge(total_return, return_count, on=['STATION', 'YEAR', 'MONTH', 'DAY', 'HOUR'], how="left" )
merge_return = merge_return.fillna(0)

merge_return.head(10)
merge_return.to_csv('./data/2013-2014_return.csv', index=False)
```
