```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('./data/2015 하반기 타슈운영현황.csv')
data = data.dropna(axis=0)

print(data)
```

```python
data['대여일시'] = pd.to_datetime(data['대여일시'].astype(str), format='%Y%m%d%H%M%S')
data['반납일시'] = pd.to_datetime(data['반납일시'].astype(str), format='%Y%m%d%H%M%S')

data['대여일시'] = data['대여일시'].dt.strftime('%Y%m%d %H')
data['반납일시'] = data['반납일시'].dt.strftime('%Y%m%d %H')
data = data.dropna(axis=0)

print(data)
```

```python
data2 = pd.read_csv('./data/2015년도 상반기 타슈운영현황.csv')

data2['대여일시'] = pd.to_datetime(data2['대여일시'].astype(str), format='%Y-%m-%d %H:%M:%S')
data2['반납일시'] = pd.to_datetime(data2['반납일시'].astype(str), format='%Y-%m-%d %H:%M:%S')

data2['대여일시'] = data2['대여일시'].dt.strftime('%Y%m%d %H')
data2['반납일시'] = data2['반납일시'].dt.strftime('%Y%m%d %H')

data2 = data2.dropna(axis=0)
print(data2)
```

```python
result = pd.concat([data2, data])
print(result)
```

```python
result = result.fillna(0)
result.to_csv('./data/2015년도 타슈현황.csv', index=False)
```
